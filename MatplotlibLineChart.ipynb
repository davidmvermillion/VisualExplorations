## Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline   # Method 1
from scipy.interpolate import interp1d   # Method 2

# matplotlib parameters
plt.rcParams['axes.axisbelow'] = True
plt.rcParams['figure.dpi'] = 300
```

## Read Data

```python
# Read data
data = pd.read_csv("yearly-number-of-objects-launched-into-outer-space.csv")
```

## Prepare Data

```python
# Subset Data
launches = data[['Year', 'yearly_launches']]

# Group by Year
launches = launches.groupby('Year', as_index = False)['yearly_launches'].sum()
launches = pd.DataFrame(launches)
```

## Default Line Plot

```python
fig1, ax = plt.subplots()
plt.plot(launches['Year'], launches['yearly_launches']);
ax.set_title('Default Line Plot of Yearly Launches');
```

## My Standard Line Plot

```python
# Method 1

# Convert data
X_Y_Spline = make_interp_spline(data['Months'], data['Sunspots'])
newx = np.linspace(data['Months'].min(), data['Months'].max(), 500)
newy = X_Y_Spline(newx)

# Plot raw data
fig1, ax = plt.subplots()
plt.plot(data['Months'], data['Sunspots'], color = '#e34234')
ax.set_title('Sunspots from 1749 and 1750', fontsize = 30).set_color('#505050')
ax.tick_params(axis='y', which='both', right=False,
                left=False, colors = '#101010')
ax.tick_params(axis='x', which='both', bottom=False, colors = '#101010')
for pos in ['right', 'top', 'left', 'bottom']:
   plt.gca().spines[pos].set_visible(False)
plt.xticks([])
plt.yticks([])

# Plot smoothed data
fig2, ax = plt.subplots()
plt.plot(newx, newy, color = '#e34234')
ax.set_title('Sunspots from 1749 and 1750', fontsize = 30).set_color('#505050')
ax.tick_params(axis='y', which='both', right=False,
                left=False, colors = '#101010')
ax.tick_params(axis='x', which='both', bottom=False, colors = '#101010')
for pos in ['right', 'top', 'left', 'bottom']:
   plt.gca().spines[pos].set_visible(False)
plt.xticks([])
plt.yticks([])

plt.show();
```

```python
# Method 2

# Convert data
cubic_interpolation_model = interp1d(data['Months'], data['Sunspots'], kind = "cubic")
newx = np.linspace(data['Months'].min(), data['Months'].max(), 500)
newy = cubic_interpolation_model(newx)

# Plot raw data
fig1, ax = plt.subplots()
plt.plot(data['Months'], data['Sunspots'], color = '#e34234')
ax.set_title('Sunspots from 1749 and 1750', fontsize = 30).set_color('#505050')
ax.tick_params(axis='y', which='both', right=False,
                left=False, colors = '#101010')
ax.tick_params(axis='x', which='both', bottom=False, colors = '#101010')
for pos in ['right', 'top', 'left', 'bottom']:
   plt.gca().spines[pos].set_visible(False)
plt.xticks([])
plt.yticks([])

# Plot smoothed data
fig2, ax = plt.subplots()
plt.plot(newx, newy, color = '#e34234')
ax.set_title('Sunspots from 1749 and 1750', fontsize = 30).set_color('#505050')
ax.tick_params(axis='y', which='both', right=False,
                left=False, colors = '#101010')
ax.tick_params(axis='x', which='both', bottom=False, colors = '#101010')
for pos in ['right', 'top', 'left', 'bottom']:
   plt.gca().spines[pos].set_visible(False)
plt.xticks([])
plt.yticks([])

plt.show();
```
