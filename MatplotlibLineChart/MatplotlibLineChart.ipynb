## Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# matplotlib parameters
plt.rcParams['axes.axisbelow'] = True
plt.rcParams['figure.dpi'] = 300
```

## Read Data

```python
# Read data
data = pd.read_csv("yearly-number-of-objects-launched-into-outer-space.csv")
```

## Prepare Data

```python
# Subset Data
launches = data[['Year', 'yearly_launches']]

# Group by Year
launches = launches.groupby('Year', as_index = False)['yearly_launches'].sum()
launches = pd.DataFrame(launches)
```

## Default Line Plot

```python
fig1, ax = plt.subplots()
plt.plot(launches['Year'], launches['yearly_launches']);
ax.set_title('Default Line Plot of Yearly Launches');
```

## My Standard Line Plot

```python
fig2, ax = plt.subplots()
plt.plot(launches['Year'], launches['yearly_launches'], color = '#e34234')
ax.set_title('Last 3 Years Showed Massive\nIncrease in Yearly Launches', fontsize = 30).set_color('#505050')
ax.set_xlabel('Year', fontsize = 15, loc =
				'left').set_color('#707070')
ax.set_ylabel('Launches', fontsize = 15, rotation =
				'horizontal', loc = 'bottom', labelpad =
				60).set_color('#707070')
for pos in ['right', 'top']:
	plt.gca().spines[pos].set_visible(False)
ax.spines['bottom'].set_color('#A0A0A0')
ax.spines['left'].set_color('#A0A0A0')
ax.tick_params(axis='y', which='both', right=False,
				left=False, colors = '#686868')
ax.tick_params(axis='x', which='both', top=False,
	       		bottom=False, colors = '#686868')
plt.show()
```
