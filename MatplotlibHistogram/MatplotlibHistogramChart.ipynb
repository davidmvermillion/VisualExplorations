## Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

# matplotlib parameters
plt.rcParams['axes.axisbelow'] = True
plt.rcParams['figure.dpi'] = 300
```

## Read Data

```python
# Read data
data = pd.read_csv("yearly-number-of-objects-launched-into-outer-space.csv")
```

## Prepare Data

```python
# Subset Data
launches = data[['Year', 'yearly_launches']]

# Group by Year
launches = launches.groupby('Year', as_index = False)['yearly_launches'].sum()
launches = pd.DataFrame(launches)
launches = launches[launches['Year'] < 2017]
```

## Default Histogram

```python
fig1, ax = plt.subplots()
plt.hist(launches['yearly_launches'], bins = 10)
ax.set_title('Default Histogram of Yearly Launches before 2017');
```

## My Standard Histogram

```python
fig2, ax = plt.subplots()
plt.hist(launches['yearly_launches'], bins = 10, color = '#2d3047')
ax.set_title('Most Years before 2017 Saw\n200 to 350 Global Launches', fontsize = 28, pad = 15).set_color('#171819')
ax.tick_params(axis = 'y', which = 'both', right = False,
				left = False, colors = '#101010')
for pos in ['right', 'top']:
	plt.gca().spines[pos].set_visible(False)
ax.spines['bottom'].set_color('#A0A0A0')
ax.spines['left'].set_color('#A0A0A0')
ax.tick_params(axis = 'x', colors = '#282828')
ax.yaxis.set_major_locator(MaxNLocator(integer=True))
ax.set_xlabel('Yearly Launches', fontsize = 16, loc = 'left').set_color('#393d3f')
ax.set_ylabel('Number\n     of\n  Years', fontsize = 16, rotation = 'horizontal', 
	      		labelpad = 65, verticalalignment = 'bottom').set_color('#393d3f')
# https://youtu.be/mmjMQkUych8
colorlist = ['#818391', '#818391', '#818391', '#818391', 
	 		'#2d3047', '#2d3047', '#2d3047',
			'#818391', '#818391', '#818391']
for i in range(len(colorlist)):
	color = colorlist[i]
	ax.patches[i].set_fc(color)

plt.show();
```
